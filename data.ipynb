```python
import pandas as pd
import ast
```

```python
data = pd.read_csv('eehc_chatbot.csv')
data.head(5)
```

```python
data.drop(columns='intent' , inplace= True)
```

```python
def extract_answer(record):
    json_dict = ast.literal_eval(record)
    return json_dict['custom']['Answer']
```

```python
data['answer'] = data['answer'].apply(extract_answer)
```

```python
data.head(5)
```

```python
data['answer'] = data['answer'].str.replace('\n', ' ')
data.head(5)
```

```python
with open('formatted_output.txt', 'w', encoding='utf-8') as file:
    for index, row in data.iterrows():
        question = row['question']
        answer = row['answer']
        file.write(f'Q: "{question}"\n')
        file.write(f'A: "{answer}"\n\n')
```
